```python
data1 = np.genfromtxt(data_path, delimiter=",", dtype=None)
print('Shape of the dataset:', data1.shape)
print('Dataset type:', type(data1))
print('A single row of the dataset is type of:', type(data1[0]))
print('Types of elements:', type(data1[0][1]), type(data1[0][3]))
print('Dataset:')
print(data1)
```

```python
dt = np.dtype("f8, f8, i8, U8")
data2 = np.genfromtxt(data_path, delimiter=",", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][3]))
print('Dataset slice:')
print(data2[:10])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Данные из отдельных столбцов
X = [] 
Y = [] 
Time = [] 

# Выполняем обход всей коллекции data2
for dot in data2:
    X.append(dot[0])
    Y.append(dot[1])
    Time.append(dot[2])

# Строим графики по проекциям данных
# Учитываем, что каждые 50 типов ирисов идут последовательно
plt.figure(1)
MISAKI, = plt.plot(X[:30], Y[:30], 'ro', label='MISAKI')
O2i3, = plt.plot(X[30:60], Y[30:60], 'g^', label='O2i3')
Lucidin, = plt.plot(X[60:90], Y[60:90], 'bs', label='Lucidin')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('X')
plt.ylabel('Y')

plt.figure(2)
MISAKI, = plt.plot(X[:30], Time[:30], 'ro', label='MISAKI')
O2i3, = plt.plot(X[30:60], Time[30:60], 'g^', label='O2i3')
Lucidin, = plt.plot(X[60:90], Time[60:90], 'bs', label='Lucidin')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('X')
plt.ylabel('Time')

plt.figure(3)
MISAKI, = plt.plot(Time[:30], Y[:30], 'ro', label='MISAKI')
O2i3, = plt.plot(Time[30:60], Y[30:60], 'g^', label='O2i3')
Lucidin, = plt.plot(Time[60:90], Y[60:90], 'bs', label='Lucidin')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Time')
plt.ylabel('Y')
plt.show()
```
